**Goal:** Analyzing the earthquake data to identify the seismic energy per country and density of the earthquakes <br/>
**Dataset**: We are investigating the Earhtquakes occured in 2021 with a magnitude greater than 4.0. The list of earthquakes is retrieved from the [Earthquakes catalog](https://www.usgs.gov/programs/earthquake-hazards/earthquakes) by USGS. Additionally, we will need the world countries and cities datasets, which are available with the geopandas module.

```python
#Importing necessary libraries
import numpy as np
import pandas as pd
import geopandas as gpd
import shapely
from shapely.geometry import Point, LineString, Polygon
import pygmt
import matplotlib.pyplot as plt
from scipy.spatial import KDTree
from scipy.spatial import cKDTree
from scipy.spatial import distance
import math
from sklearn.neighbors import KernelDensity
import seaborn
```

```python
#Loading in the Earthquake file and World countries
world_countries = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
USGS_EQ = pd.read_csv('USGS_Global_Earthquakes_2021_GT_4Mag.csv')
world_countries = world_countries[['name', 'geometry']]
USGS_EQ
```

At the moment the earthquake dataset is loaded as a pandas dataframe, however, to make spatial analyses, we need to transform it into a Geopandas dataframe. A GPD dataframe requires column to identify the geometry.<br/>
Therefore, we define a simple function in the cell below, to bring the longitude, latitude, and the depth of earthquake into a single geometry column.

```python
points = np.empty(len(USGS_EQ), object)
def pointy(df):
    for i in range(len(df)):
      
        points[i] = Point(df.loc[i].longitude,df.loc[i].latitude,df.loc[i].depth )
    return points   
USGS_EQ['geometry'] = pointy(USGS_EQ)
```

Now, the dataframe is ready to be transformed into a Geopandas dataframe, however, we need to use a Coordinate reference system. We chose *EPSG:4326* which is a Geographic CRS that uses degree as the unit. More about CRS could be found [here](https://epsg.io/). However, for our analysis, we will use a second CRS that employs metric system, which is *EPSG:3395*.

```python
#Setting a CRS and transfroming into GPD
crs = 'epsg:4326'
USGS_EQ_gdf = gpd.GeoDataFrame(USGS_EQ, crs=crs, geometry=USGS_EQ['geometry'])
USGS_EQ_gdf.head()
```

To bring the earthquake data and the list of countries, we perform a spatial join based on the Geometry. 

```python
countries_EQ = world_countries.sjoin(USGS_EQ_gdf[['mag', 'geometry', 'depth', 'longitude', 'latitude']], how="left", predicate="intersects")
# Removing NaN values
countries_EQ_cleaned = countries_EQ.dropna(subset=['mag','depth','longitude','latitude'])
countries_EQ_cleaned
```

We perform another spatial join and set the CRS *3395*. We'll need the result of this join later in the notebook.

```python
EQ_countries = USGS_EQ_gdf[['mag', 'geometry', 'depth', 'longitude', 'latitude']].sjoin(world_countries, how="left", predicate="intersects")
EQ_countries = EQ_countries.to_crs('EPSG:3395')
# Removing NaN values
EQ_countries = EQ_countries.dropna(subset=['name'])
EQ_countries
```

```python
distinct_values = list(set(countries_EQ_cleaned.index))
print(len(distinct_values))
```

We want to aggregate all earthquakes to define seismic energy per country. Therefore, a function that makes the logarithmic sum is defined in the cell below. 

```python
def sum_magnitude(df):
    mag_sum = np.log10(np.sum((10**df['mag'])))

    return(mag_sum)
```

Additionally, the mean depth of the eartquakes per country is also calculated.

```python
mean_depth = countries_EQ_cleaned.groupby(['name']).mean()['depth']
mean_depth
```

```python
magnitude_sum = countries_EQ_cleaned.groupby(['name']).apply(sum_magnitude).sort_values(ascending=False)
magnitude_sum[magnitude_sum < 0] = 0
magnitude_sum = pd.Series(magnitude_sum,name='mag_sum')
magnitude_sum
```

## Calculating mean 2D distance between Earthquakes

```python
def mean_distance_to_other_points(geodataframe):
    mean_distances = []
    for index, reference_point in geodataframe.iterrows():
        distances = []
        for _, row in geodataframe.iterrows():
            distance = math.sqrt((row.geometry.x - reference_point.geometry.x)**2 + (row.geometry.y - reference_point.geometry.y)**2)

            distances.append(distance)
        mean_distance = np.mean(distances)
        mean_distances.append(mean_distance)
    return np.mean(mean_distances)
```

```python
# Calculating the mean distance for the all earthquakes, however, this may take a little long
mean_2D_distance = np.mean(mean_distance_to_other_points(EQ_countries))
```

## Aggregated seismic magnitude for cities

We first retrieve the citites dataset and its crs to *3395*, as km is a more appropriate unit for our analysis. Here, our goal is to create aggregated seismic magnitudes for the cities. However, in contrast the *countries* dataset, the cities are defined by a single point. Therefore, to find the relevant earthquakes for each city we implement a different approach. <br\>
We use KDTrees, since they enable query the points with a given radius.

```python
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
print(cities.crs)
cities = cities.to_crs('EPSG:3395')
print(cities.crs)
cities
```

```python
cities['lon'] = cities.geometry.x
cities['lat'] = cities.geometry.y
cities
```

```python
# KDTrees are generated from an array. Therefore, we create arrays for the cities and earthquakes.
USGS_EQ_gdf_km = USGS_EQ_gdf.to_crs('EPSG:3395')
cities_arr = np.zeros(shape = (len(cities),2))
cities_arr[:,0] = cities.lon 
cities_arr[:,1] = cities.lat 

quakes_arr = np.zeros(shape=(len(USGS_EQ_gdf),2))
quakes_arr[:,0] = USGS_EQ_gdf_km.geometry.x
quakes_arr[:,1] = USGS_EQ_gdf_km.geometry.y
```

```python
quakes_tree = cKDTree(quakes_arr)
cities_tree = cKDTree(cities_arr)
```

We are searching for all the earthquakes within 50 km radius for each city.

```python
r = 50000 # The unit for our CRS is meters.
idx = quakes_tree.query_ball_point(cities_arr, r=r, return_sorted=True)
```

The query return us a list with the length of 202 (the same as the length of city dataset) which contains the indices of corresponding earthquakes as list.

Now, we have the earthquake information corresponding each city. However, how we can use that? We need to somehow *combine* this information with our datasets. In other words, first we need to merge or join the result of query to cities dataset. Because we can use the index of *idx* and index of cities df as shared column for our operation. Following that we will have a df containing cities and their earthquakes' indexes. Therefore, we can use these indexes to join the last df to the USGS df to have the city names and their relevant earthquake informations.

```python
index = np.arange(0,202,1)

bos_df = pd.DataFrame(columns={'idc'},index=index)

for i in range(len(idx)):
    if(len(idx[i])>0):
        m = len(idx[i])
        if (m>0):
            a = np.zeros(shape = (m))
            
            for j in range(m):

                a[j] = idx[i][j]

            bos = pd.DataFrame(data={'idc':a}, index = [i for k in range(len(a))])
            bos_df = pd.concat([bos_df, bos])

bos_df = bos_df.dropna(axis=0)            
bos_df
```

```python
bos_df['arg'] = bos_df.index
cities['arg'] = np.arange(0,202,1)
```

```python
df_left = pd.merge(cities, bos_df, on='arg', how='left')
df_left
```

```python
USGS_EQ_gdf['idc'] = USGS_EQ_gdf.index
cities_EQ = pd.merge(df_left[['name','geometry','idc']], USGS_EQ_gdf[['mag', 'geometry', 'depth', 'longitude', 'latitude','idc']], on='idc', how='left')
cities_EQ = cities_EQ.dropna(subset=['mag'])
cities_EQ
```

For the next part of question (aggregated seismic magnitude), I applied the function I defined for the first question.

```python
city_mag_sum = cities_EQ.groupby(['name']).apply(sum_magnitude).sort_values(ascending=False)
city_mag_sum[city_mag_sum < 0] = np.nan
city_mag_sum = pd.Series(city_mag_sum, name='agg_sum')
```

Let us merge it to the cities dataframe.

```python
cities_mag_sum = cities.merge(city_mag_sum, on = 'name').sort_values('agg_sum', ascending=False)
cities_mag_sum
```
